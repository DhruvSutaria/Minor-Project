```python
import re    
import nltk   
import string 
import warnings 
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("sentiments.csv")
```

```python
df
```

```python
DATASET_COLUMNS = ["target", "ids", "date", "flag", "user", "Tweets"]
df.columns = DATASET_COLUMNS
df.head()
```

```python
df.drop(['ids','date','flag','user'],axis = 1,inplace = True)
```

```python
df
```

```python
df['Clean_Tweets'] = df['Tweets'].str.replace("@", "") 
df.head()
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].str.replace(r"http\S+", "") 
df.head()
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].str.replace("[^a-zA-Z]", " ") 
df.head()
```

```python
nltk.download('stopwords')
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    clean_text=' '.join([word for word in text.split() if word not in stopwords])
    return clean_text
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].apply(lambda text : remove_stopwords(text.lower()))
df.head()
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].apply(lambda x: x.split())
df.head()
```

```python
from nltk.stem.porter import * 
stemmer = PorterStemmer() 
df['Clean_Tweets'] = df['Clean_Tweets'].apply(lambda x: [stemmer.stem(i) for i in x])
```

```python
df.head()
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].apply(lambda x: ' '.join([w for w in x]))
df.head()
```

```python
df['Clean_Tweets'] = df['Clean_Tweets'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
df.head()
```
